Q1 What is 2 tailed and  tailed T Test ?

# T-Test Explanation

# A T-test is a statistical test used to compare means (averages)
# and check whether the difference between them is significant or just due to chance.

# 1) One-Tailed T-Test:
# A one-tailed test checks for a difference in ONE direction only.
# It tests whether a value is either greater than OR less than another value.

# Example:
# H0 (Null Hypothesis): Mean = 50
# H1 (Alternative Hypothesis): Mean > 50  (Right-tailed)
# OR
# H1: Mean < 50  (Left-tailed)

# Use one-tailed test when you are interested in only one direction.

# 2) Two-Tailed T-Test:
# A two-tailed test checks for a difference in BOTH directions.
# It tests whether a value is either greater than OR less than another value.

# Example:
# H0: Mean = 50
# H1: Mean ≠ 50

# Use two-tailed test when you want to check any difference (increase or decrease).

# Key Difference:
# One-tailed → Tests one direction (>, <)
# Two-tailed → Tests both directions (≠)

# Conclusion:
# Choose one-tailed if direction is known.
### Choose two-tailed if checking for any change.

**Q3.What is power of statistical test?**


# Power of a Statistical Test

# Power of a test is the probability of correctly rejecting the null hypothesis (H0)
# when it is actually FALSE.

# In simple words:
# Power tells us how good a test is at detecting a real effect.

# Formula:
# Power = 1 - β
# where β (beta) = probability of Type II error

# Meaning:
# High power → Less chance of missing a real effect
# Low power → More chance of missing a real effect

# Example:
# H0: A new medicine has no effect
# If the medicine actually works and the test correctly detects it,
# then this is due to high power of the test.

# If the test fails to detect that the medicine works,
# it means low power (Type II error occurred).

# Key Points:
# Power ranges from 0 to 1
# Higher power (e.g., 0.8 or 80%) is considered good

# Factors affecting power:
# 1) Sample size → Larger sample = Higher power
# 2) Effect size → Bigger difference = Higher power
# 3) Significance level (alpha) → Higher alpha = Higher power
# 4) Variability → Less variation = Higher power

# Conclusion:
# Power helps us understand how reliable a test is in detecting real differences.

**Q4  Explain input and out put characteristics for  stats.ttest_rel function in your own words**


# stats.ttest_rel() Function (Paired T-Test)

# This function is used to compare TWO RELATED samples.
# It checks whether the mean difference between paired observations is significant.

# Example of related samples:
# - Before and after exam marks of same students
# - Weight of same person before and after diet

# INPUT CHARACTERISTICS:

# 1) Two datasets (a, b):
# These are paired data (same length).
# Example:
# a = marks before training
# b = marks after training

# 2) axis (optional):
# Defines how the data is arranged (row-wise or column-wise).

# 3) nan_policy (optional):
# How to handle missing values (NaN)
# 'propagate' → return NaN
# 'omit' → ignore NaN values
# 'raise' → give error

# 4) alternative (optional):
# Defines the type of test:
# 'two-sided' → checks for any difference
# 'less' → checks if a < b
# 'greater' → checks if a > b


# OUTPUT CHARACTERISTICS:

# The function returns TWO main values:

# 1) t-statistic:
# This tells how much difference exists between the two datasets.
# Higher absolute value → bigger difference

# 2) p-value:
# This tells whether the difference is statistically significant.

# Interpretation:
# If p-value < 0.05 → significant difference (reject H0)
# If p-value >= 0.05 → no significant difference (fail to reject H0)

# Example usage:
from scipy import stats

a = [10, 12, 14, 16, 18]
b = [11, 13, 15, 17, 19]

result = stats.ttest_rel(a, b)

print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)

# Conclusion:
# stats.ttest_rel() is useful when comparing related/paired data.

**Q5 What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance**

# Function to Identify Power of a Statistical Test

# In Python, the 'scipy.stats' module does NOT directly provide a function
# to calculate statistical power.

# For power calculation, we generally use:
# statsmodels.stats.power module

# Common functions:
# 1) TTestPower()
# 2) TTestIndPower()
# 3) NormalIndPower()

from statsmodels.stats.power import TTestPower

# Example:
analysis = TTestPower()

# Calculate power:
power = analysis.power(effect_size=0.5, nobs=30, alpha=0.05)

print("Power of test:", power)


# OUTPUT RANGE:

# Power value always lies between:
# 0 to 1

# Meaning:
# 0   → No ability to detect effect
# 1   → Perfect ability to detect effect

# Common interpretation:
# Power = 0.8 (80%) is considered good
# This means there is 80% chance of detecting a real effect


# SIGNIFICANCE OF POWER:

# 1) Helps avoid Type II Error:
# Higher power → Lower chance of missing a real effect

# 2) Improves reliability:
# High power means results are more trustworthy

# 3) Helps in sample size planning:
# We can decide how much data is needed before experiment

# 4) Better decision making:
# Ensures we correctly reject false null hypothesis (H0)

# Conclusion:
# Power is an important measure to check how effective a statistical test is.
