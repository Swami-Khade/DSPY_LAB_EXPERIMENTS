**Q1 When do we go for Chi square test**?  

# Chi-Square Test (χ² Test)

# The Chi-square test is used to check relationships between CATEGORICAL data.

# WHEN DO WE USE CHI-SQUARE TEST?

# 1) When data is categorical (not numerical)
# Example:
# Gender (Male/Female), Yes/No, Pass/Fail

# 2) To test independence between two variables
# Example:
# Is there a relationship between Gender and Product Choice?

# 3) To compare observed vs expected frequencies
# Example:
# Did actual results match expected results?

# 4) When sample size is large
# (Expected frequency in each category should be at least 5)

# TYPES OF CHI-SQUARE TEST:

# 1) Chi-square Test for Independence:
# Checks whether two categorical variables are related

# Example:
# H0: Gender and product choice are independent
# H1: They are related

# 2) Chi-square Goodness of Fit Test:
# Checks whether observed data fits expected distribution

# Example:
# H0: Dice is fair
# H1: Dice is not fair


# KEY POINTS:

# - Works only with frequency (count) data
# - Not suitable for small sample sizes
# - Does not tell strength of relationship, only existence

# CONCLUSION:
# Use Chi-square test when dealing with categorical data
# and you want to check relationship or distribution.

**Q2 Do we use Chi square test for independence? Explain How?**

# Chi-Square Test for Independence

# Yes, Chi-square test is used to check independence
# between two categorical variables.

# PURPOSE:
# To find whether two variables are related or independent

# Example:
# Gender vs Product Choice

# HYPOTHESIS:
# H0: Variables are independent (no relation)
# H1: Variables are dependent (related)

# HOW IT WORKS:

# 1) Create contingency table (observed values)
# Example:
#        Product A   Product B
# Male       20          30
# Female     25          25

# 2) Calculate expected values using formula:
# Expected = (Row Total × Column Total) / Grand Total

# 3) Apply Chi-square formula:
# χ² = Σ [(Observed - Expected)² / Expected]

# 4) Find p-value

# DECISION:
# If p-value < 0.05 → Reject H0 (variables are related)
# If p-value ≥ 0.05 → Accept H0 (variables are independent)

# CONCLUSION:
# Chi-square test helps us check relationship between categorical variables

**Q3 Explain role of Critical value and p-value in Chi square test**

# Role of Critical Value and p-value in Chi-Square Test

# 1) Critical Value:
# It is a threshold value taken from Chi-square table
# based on significance level (alpha) and degrees of freedom

# Role:
# Compare calculated χ² value with critical value

# Decision:
# If χ² > Critical Value → Reject H0
# If χ² ≤ Critical Value → Accept H0


# 2) p-value:
# It shows probability of getting result by chance

# Role:
# Directly tells significance of result

# Decision:
# If p-value < 0.05 → Reject H0
# If p-value ≥ 0.05 → Accept H0


# Conclusion:
# Both methods help decide whether variables are related or independent
# p-value method is more commonly used

**Q4 Can we use chi square test to show existence of correlation between 2 features in a dataset?  How will you interpret test result to indicate correlation presenceर**

# Chi-Square Test and Correlation

# Yes, Chi-square test can be used to check RELATION (association)
# between two categorical variables (not exact correlation value).

# It does NOT give strength/direction like correlation coefficient,
# only tells whether relation exists or not.

# HYPOTHESIS:
# H0: Variables are independent (no relation)
# H1: Variables are dependent (relation exists)

# INTERPRETATION:

# Step 1: Perform Chi-square test → get p-value

# Step 2: Decision:
# If p-value < 0.05:
# → Reject H0
# → Conclusion: Relationship exists (correlation present)

# If p-value ≥ 0.05:
# → Fail to reject H0
# → Conclusion: No relationship (no correlation)

# IMPORTANT:
# Chi-square only shows existence of association,
# NOT how strong or positive/negative it is.

**Q5 What is degrees of freedom and how to calculate it while performing chi square test**

# Degrees of Freedom (DoF) in Chi-Square Test

# Degrees of freedom means number of values that are free to vary
# after some constraints are applied.

# In simple words:
# It tells how much independent information is available.

# FORMULA (for Chi-square test of independence):

# DoF = (rows - 1) × (columns - 1)

# Example:
# If table has 3 rows and 2 columns:
# DoF = (3 - 1) × (2 - 1) = 2 × 1 = 2

# ROLE:
# Degrees of freedom is used to find critical value
# from Chi-square table.

# CONCLUSION:
# DoF helps in deciding correct threshold for hypothesis testing.
