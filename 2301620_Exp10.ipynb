**Q1 Explain input and output characteristics for f_oneway function from stats library**


# stats.f_oneway() Function (One-Way ANOVA)

# This function is used to compare means of 3 or more groups
# to check if at least one group is different.

# INPUT CHARACTERISTICS:

# 1) Multiple datasets (2 or more groups)
# Example:
# group1 = [10, 12, 14]
# group2 = [15, 18, 20]
# group3 = [22, 24, 26]

# 2) Data should be numerical

# 3) Groups should be independent

# 4) Optional parameters:
# axis → defines data arrangement
# nan_policy → handling missing values


# OUTPUT CHARACTERISTICS:

# The function returns TWO values:

# 1) F-statistic:
# Measures variation between groups
# Higher value → more difference

# 2) p-value:
# Shows significance of difference

# INTERPRETATION:

# If p-value < 0.05:
# → Reject H0 → At least one group mean is different

# If p-value ≥ 0.05:
# → Fail to reject H0 → No significant difference


# EXAMPLE:
from scipy import stats

group1 = [10, 12, 14]
group2 = [15, 18, 20]
group3 = [22, 24, 26]

result = stats.f_oneway(group1, group2, group3)

print("F-statistic:", result.statistic)
print("p-value:", result.pvalue)

# CONCLUSION:
# f_oneway() helps compare multiple group means

**Q2 Explain results of sm.stats.anova_lm fucntion**

# sm.stats.anova_lm() Function (ANOVA Table)

# This function (from statsmodels) is used to perform ANOVA
# on a fitted linear model and gives results in table format.

# OUTPUT CHARACTERISTICS:

# The result is a table with following columns:

# 1) sum_sq (Sum of Squares):
# Measures total variation for each factor

# 2) df (Degrees of Freedom):
# Number of independent values for each factor

# 3) F (F-statistic):
# Ratio of variance → shows difference between groups

# 4) PR(>F) (p-value):
# Probability value to test significance

# INTERPRETATION:

# Check p-value (PR(>F)):

# If p-value < 0.05:
# → Factor is significant (affects result)

# If p-value ≥ 0.05:
# → Factor is NOT significant

# MODEL ROWS:

# - Each row represents a variable (factor)
# - Residual row shows error/unexplained variation

# CONCLUSION:
# anova_lm() helps identify which variables
# significantly affect the dependent variable

**Q3 What role ANOVA plays in data analytics?**


# Role of ANOVA in Data Analytics

# ANOVA (Analysis of Variance) is used to compare means
# of 3 or more groups.

# ROLE IN DATA ANALYTICS:

# 1) Identify Differences:
# Helps find if different groups have different averages

# 2) Feature Importance:
# Shows which factors significantly affect outcome

# 3) Decision Making:
# Helps choose best option among multiple groups

# 4) Hypothesis Testing:
# Tests whether variation is real or due to chance

# 5) Model Building:
# Used in regression and experimental analysis

# INTERPRETATION:

# If p-value < 0.05:
# → Significant difference exists

# If p-value ≥ 0.05:
# → No significant difference

# ANOVA helps analyze and compare multiple groups effectively

**Q4 When to use one way ANOVA? Explain with real world case**

# One-Way ANOVA

# WHEN TO USE:
# Use one-way ANOVA when:
# 1) Comparing means of 3 or more groups
# 2) Only ONE independent variable (factor)
# 3) Data is numerical and groups are independent

# REAL-WORLD EXAMPLE:

# Problem:
# A teacher wants to check if teaching method affects marks

# Groups:
# Method A, Method B, Method C

# Marks:
# A = [70, 75, 80]
# B = [85, 88, 90]
# C = [60, 65, 70]

# HYPOTHESIS:
# H0: All group means are equal
# H1: At least one group mean is different

# INTERPRETATION:

# If p-value < 0.05:
# → Teaching method affects marks

# If p-value ≥ 0.05:
# → No effect of teaching method

# CONCLUSION:
# One-way ANOVA helps compare multiple groups using one factor

**Q5 When to use 2 way ANOVA ? Explain with real world case.**


# Two-Way ANOVA

# WHEN TO USE:
# Use two-way ANOVA when:
# 1) There are TWO independent variables (factors)
# 2) You want to study their individual and combined effect
# 3) Dependent variable is numerical

# REAL-WORLD EXAMPLE:

# Problem:
# A company wants to check effect of:
# 1) Study Method (Online / Offline)
# 2) Study Time (Morning / Evening)
# on student marks

# Factors:
# Factor 1 → Method
# Factor 2 → Time

# HYPOTHESIS:
# H0: No effect of method, time, or interaction
# H1: At least one effect is significant

# WHAT IT CHECKS:
# 1) Effect of Method alone
# 2) Effect of Time alone
# 3) Interaction effect (Method × Time)

# INTERPRETATION:

# If p-value < 0.05 for any factor:
# → That factor has significant effect

# CONCLUSION:
# Two-way ANOVA helps analyze two factors and their interaction
